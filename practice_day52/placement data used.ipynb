```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
class mera2LR:
    def __init__(self):
        self.m = None
        self.b = None

    def fit(self, X_train, y_train):
        num1 = 0
        den1 = 0

        for i in range(X_train.shape[0]):
            num1 += ((X_train[i] - X_train.mean()) * (y_train[i] - y_train.mean()))
            den1 += ((X_train[i] - X_train.mean()) * (X_train[i] - X_train.mean()))


        self.m = num1 / den1
        self.b = (y_train.mean() - (self.m * X_train.mean()))

    def predict(self, X_test):
        return self.m * X_test + self.b


    def MAE(self, X_test, y_test):

        MAEValue = 0
        
        y_pred1 = self.predict(X_test)
        
        for j in range(len(y_test)):
            MAEValue +=  abs(y_test[j] - y_pred1[j])
    
        return MAEValue / len(y_test)


    def MSE(self, X_test, y_test):
        MSEValue = 0
        y_pred2 = self.predict(X_test)

        for k in range(len(y_test)):
            MSEValue += ((y_test[k] - y_pred2[k]) * (y_test[k] - y_pred2[k]))

        return MSEValue / len(y_test)



    def RMSE(self, X_test, y_test):

        RMSEValue = 0

        y_pred3 = self.predict(X_test)

        for l in range(len(y_test)):
            RMSEValue += ((y_test[l] - y_pred3[l]) * (y_test[l] - y_pred3[l]))

        return np.sqrt(RMSEValue / len(y_test))


    def R2(self, X_test, y_test):
        
        R2Value = 0

        y_pred4 = self.predict(X_test)

        num2 = 0
        den2 = 0

        for m in range(len(y_test)):
            num2 += ((y_test[m] - y_pred4[m]) * (y_test[m] - y_pred4[m]))
            den2 += ((y_test[m] - y_test.mean()) * (y_test[m] - y_test.mean()))


        return (1 - (num2 / den2))
        

    def AdjR2(self, X_test, y_test):
        n = len(X_test)
        k = p = X_test.shape[1] if X_test.ndim > 1 else 1


        return (1 - ((1 - self.R2(X_test, y_test)) * (n - 1)) / (n - 1 - k))
```
