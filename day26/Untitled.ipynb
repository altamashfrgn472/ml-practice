```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

# Example data
X_train = [['Poor', 'School'], ['Average', 'UG'], ['Good', 'PG']]
X_test = [['Poor', 'School'], ['Average', 'UG'], ['Good', 'PG']]

# Convert to DataFrame with feature names
X_train = pd.DataFrame(X_train, columns=['review', 'education'])
X_test = pd.DataFrame(X_test, columns=['review', 'education'])

# Initialize OrdinalEncoder
oe = OrdinalEncoder(categories=[['Poor', 'Average', 'Good'], ['School', 'UG', 'PG']])

# Fit the encoder
oe.fit(X_train)

# Transform the data
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)

print("Encoded X_train:")
print(X_train)
print("Encoded X_test:")
print(X_test)
```
