```python
import pandas as pd
import requests

all_data = []  # List to collect all data frames

for i in range(1, 497):
    response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=cc7ae4e04690c3e9fdb1d499a17a90d6&languagetmdb=en-US&page={i}')
    
    if response.status_code == 200:
        # Parse response and create a DataFrame
        data = response.json()
        if 'results' in data and data['results']:
            temp_df = pd.DataFrame(data['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
            all_data.append(temp_df)  # Add the DataFrame to the list
        else:
            print(f"No results found at page {i}. Stopping.")
            break
    else:
        print(f"Error fetching data for page {i}, Status Code: {response.status_code}")
        break

# Combine all collected DataFrames into a single DataFrame
df = pd.concat(all_data, ignore_index=True)

# Preview the resulting DataFrame
print(df.head())
```
